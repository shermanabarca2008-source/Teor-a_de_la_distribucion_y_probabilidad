---

## 🎯 Planteamiento del problema

Se plantea la siguiente pregunta de investigación:

> **¿La remuneración mensual promedio de los servidores públicos registrados en este conjunto de datos de Loja es igual a 700 dólares estadounidenses?**

Considerando que el conjunto de datos está conformado por una **muestra de 2883 observaciones (n = 2883)** y que **no se conoce la desviación estándar poblacional**, la prueba estadística más apropiada para este análisis es la **Prueba *t* de Student para una muestra**.

Con base en este planteamiento, se formulan las siguientes hipótesis.

---

# 📌 Formulación de Hipótesis

### 🔹 Hipótesis nula (\(H_0\))

> $$H_0: \mu = 700$$

La **remuneración mensual promedio** de los servidores públicos es igual a **700 USD**.

---

### 🔹 Hipótesis alternativa (\(H_1\))

> $$H_1: \mu \neq 700$$

La **remuneración mensual promedio** de los servidores públicos es diferente de **700 USD**.

---

## 📊 Nivel de significancia

Para la prueba de hipótesis se establece un nivel de significancia de:

$$\alpha = 0.05$$

Este valor representa una **probabilidad del 5 % de cometer un error de Tipo I**, es decir, rechazar la hipótesis nula cuando en realidad es verdadera.

---

## ✅ Resumen

| Concepto | Descripción |
|----------|-------------|
| 📌 **Hipótesis nula (\(H_0\))** | $ \mu = 700 $ |
| 📌 **Hipótesis alternativa (\(H_1\))** | $ \mu \neq 700 $ |
| 📊 **Prueba estadística** | *t* de Student para una muestra |
| 🎯 **Nivel de significancia** | $ \alpha = 0.05 $ (5 %) |
| 👥 **Tamaño de la muestra** | $ n = 2883 $ |

---

Una vez ejecutada la prueba *t* de Student, la decisión estadística se tomará a partir del **valor-p (p-value)** obtenido. Si el valor-p es menor que el nivel de significancia establecido ($ \alpha = 0.05 $), se rechazará la hipótesis nula; en caso contrario, no existirá evidencia estadística suficiente para rechazarla.

```python
import pandas as pd
from scipy.stats import ttest_1samp

df = pd.read_excel("3._remuneraciones_ingresos_adicionales.xlsx")

salarios = df["Remuneración mensual unificada"].dropna()

# Hipótesis
# H0: μ = 700
# H1: μ ≠ 700

t, p = ttest_1samp(salarios, popmean=700)

print("Estadístico T:", t)
print("Valor-p:", p)
```



## 📊 Comparación de Grupos

Para resolver esta interrogante, se ha utilizado la variable **Remuneración Mensual Unificada**, debido a que es una **variable cuantitativa**. Como variable independiente se ha seleccionado el **Régimen Laboral**, planteando la siguiente pregunta de investigación:

> **¿Existen diferencias significativas entre la remuneración mensual promedio de los diferentes regímenes laborales del sector público en Loja?**

A partir de este planteamiento, es posible realizar la comparación entre los distintos grupos presentes en el conjunto de datos, aplicando un **Análisis de Varianza (ANOVA de un factor)** y, posteriormente, una **prueba Post-Hoc de Tukey**, con el fin de determinar si existen diferencias estadísticamente significativas entre las medias de remuneración mensual correspondientes a cada régimen laboral.

```python
# Autor: Sherman Abarca
# IMPORTAR LIBRERÍAS


import pandas as pd
import numpy as np

from scipy.stats import shapiro
from scipy.stats import levene
from scipy.stats import f_oneway

from statsmodels.stats.multicomp import pairwise_tukeyhsd

# CARGAR DATASET

df = pd.read_excel("3._remuneraciones_ingresos_adicionales.xlsx")

print("="*60)
print("PRIMERAS FILAS DEL DATASET")
print("="*60)

print(df.head())

# INFORMACIÓN GENERAL

print("\n")
print("="*60)
print("INFORMACIÓN DEL DATASET")
print("="*60)

print(df.info())

# ESTADÍSTICAS DESCRIPTIVAS

print("\n")
print("="*60)
print("ESTADÍSTICAS DESCRIPTIVAS")
print("="*60)

print(df["Remuneración mensual unificada"].describe())

# CANTIDAD DE REGISTROS POR GRUPO

print("\n")
print("="*60)
print("REGÍMENES LABORALES")
print("="*60)

print(df["Régimen laboral al que pertenece\xa0"].value_counts())

# ELIMINAR DATOS NULOS

datos = df[["Régimen laboral al que pertenece\xa0",
            "Remuneración mensual unificada"]].dropna()

# CREAR LOS GRUPOS

grupos = []

print("\n")
print("="*60)
print("PRUEBA DE NORMALIDAD (SHAPIRO)")
print("="*60)

for nombre, grupo in datos.groupby("Régimen laboral al que pertenece\xa0"):

    salarios = grupo["Remuneración mensual unificada"]

    grupos.append(salarios)

    # Shapiro recomienda muestras <=5000
    muestra = salarios

    if len(muestra) > 5000:
        muestra = muestra.sample(5000, random_state=1)

    stat, p = shapiro(muestra)

    print(f"\nGrupo: {nombre}")
    print("Cantidad:", len(salarios))
    print("p =", round(p,5))

    if p > 0.05:
        print("Distribución aproximadamente normal")
    else:
        print("Distribución NO normal")

# HOMOGENEIDAD DE VARIANZAS

print("\n")
print("="*60)
print("PRUEBA DE LEVENE")
print("="*60)

estadistico, p = levene(*grupos)

print("Estadístico:", estadistico)
print("Valor-p:", p)

if p > 0.05:
    print("Las varianzas pueden considerarse iguales.")
else:
    print("Las varianzas son significativamente diferentes.")

# ANOVA

print("\n")
print("="*60)
print("ANOVA DE UN FACTOR")
print("="*60)

F, p = f_oneway(*grupos)

print("Estadístico F:", round(F,4))
print("Valor-p:", p)

if p < 0.05:

    print("\nSe RECHAZA la hipótesis nula.")
    print("Existen diferencias significativas entre las medias.")

else:

    print("\nNo se rechaza la hipótesis nula.")
    print("No existen diferencias significativas.")

# PRUEBA POST-HOC DE TUKEY

print("\n")
print("="*60)
print("PRUEBA POST-HOC DE TUKEY")
print("="*60)

tukey = pairwise_tukeyhsd(

    endog=datos["Remuneración mensual unificada"],

    groups=datos["Régimen laboral al que pertenece\xa0"],

    alpha=0.05

)

print(tukey)

# TABLA DE RESULTADOS

tabla = pd.DataFrame(

    data=tukey._results_table.data[1:],

    columns=tukey._results_table.data[0]

)

print("\n")
print("="*60)
print("TABLA DE TUKEY")
print("="*60)

print(tabla)

# INTERPRETACIÓN AUTOMÁTICA

print("\n")
print("="*60)
print("INTERPRETACIÓN")
print("="*60)

for i in range(len(tabla)):

    g1 = tabla.loc[i,"group1"]
    g2 = tabla.loc[i,"group2"]
    pvalor = tabla.loc[i,"p-adj"]
    rechazo = tabla.loc[i,"reject"]

    if rechazo:

        print(f"Entre {g1} y {g2} SI existen diferencias significativas (p={pvalor}).")

    else:

        print(f"Entre {g1} y {g2} NO existen diferencias significativas (p={pvalor}).")

# CONCLUSIÓN FINAL

print("\n")
print("="*60)
print("CONCLUSIÓN")
print("="*60)

if p < 0.05:

    print("""
El análisis ANOVA indica que existen diferencias estadísticamente
significativas entre las remuneraciones mensuales promedio de al
menos dos regímenes laborales.

La prueba Post-Hoc de Tukey permite identificar exactamente qué
pares de grupos presentan dichas diferencias.
""")

else:

    print("""
El análisis ANOVA no encontró diferencias estadísticamente
significativas entre las medias de los regímenes laborales.
""")
```

## 📊 Prueba Post-Hoc de Tukey

La **prueba Post-Hoc de Tukey** evidenció diferencias estadísticamente significativas entre los regímenes laborales **LOSEP** y **Código de Trabajo** (\(p < 0.05\)), lo que indica que las **remuneraciones mensuales promedio** de ambos grupos son significativamente diferentes.

Por otra parte, al comparar los regímenes **LOSEP** y **Contrato**, no se encontraron diferencias estadísticamente significativas (\(p > 0.05\)). En consecuencia, no existe evidencia suficiente para afirmar que las remuneraciones mensuales promedio de estos dos grupos sean diferentes.

> **Conclusión**
>
> - ✅ Existe una diferencia significativa entre **LOSEP** y **Código de Trabajo**.
> - ❌ No existe una diferencia significativa entre **LOSEP** y **Contrato**.
> - 📈 Los resultados de la prueba de Tukey permiten identificar específicamente entre qué grupos se presentan diferencias en la remuneración mensual promedio.

## 📌 Conclusión de los Resultados

Se aplicó un **Análisis de Varianza (ANOVA de un factor)** con el propósito de comparar la **remuneración mensual unificada** entre los diferentes **regímenes laborales** presentes en el conjunto de datos correspondiente a la ciudad de Loja.

El análisis permitió determinar si las diferencias observadas entre las medias de los grupos eran **estadísticamente significativas**. Posteriormente, se aplicó la **prueba Post-Hoc de Tukey**, la cual permitió identificar de manera específica qué pares de regímenes laborales presentaban diferencias significativas en sus remuneraciones mensuales promedio.

En conjunto, estos procedimientos estadísticos proporcionan un análisis más detallado y confiable de las diferencias salariales existentes entre los distintos regímenes laborales, facilitando una mejor interpretación de los resultados obtenidos y respaldando las conclusiones derivadas del estudio.
